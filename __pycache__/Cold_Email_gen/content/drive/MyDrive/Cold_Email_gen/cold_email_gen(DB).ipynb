```python
import chromadb

client = chromadb.Client()

try:
    # Attempt to get the collection, it will create it if it doesn't exist
    collection = client.get_or_create_collection("my_collection")
except chromadb.errors.UniqueConstraintError:
    # If it exists, retrieve it instead
    collection = client.get_collection("my_collection")

```

```python
# Fetch existing IDs from the collection
existing_ids = collection.get()['ids']

# Define new IDs
new_ids = ["id1", "id2"]

# Filter out duplicate IDs
unique_ids = [id_ for id_ in new_ids if id_ not in existing_ids]

# Add only unique IDs
if unique_ids:
    collection.add(
        documents=["This is about NewYork ",
                   "This is another document about Delhi"],
        ids=unique_ids,
        metadatas=[{"url": "https://en.wikipedia.org/wiki/New_York_City"},
                   {"url": "https://en.wikipedia.org/wiki/New_Delhi"}]
    )
```

```python
all_docs = collection.get()
print(all_docs)
```

```python
collection.query(
    query_texts=["This is about Kareena KApoor"],
    n_results=2
)
```
